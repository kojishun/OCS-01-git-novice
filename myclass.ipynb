### 授業の目標と概要
情報系以外を専門とする層にとって，コンピュータシステムの知識を体系的に学ぶことはリソースの観点から難しいのが現状である．そこで，本授業では，必要なコンピュータシステムの知識を比較的短い期間で効率的に学ぶことを目標としている．授業は反転授業に近い方式で行い，実践を通じた学習によって知識とスキルの獲得を目指し，コンピュータシステムを用いた課題解決を実際に行えるようになることを重視する．


>コンピュータシステムを利用した情報サービスの知識はあらゆる分野で求められている。
本講義では、情報サービスの提供に必要な知識・スキルに加えてそれらの獲得方法を学ぶ。
具体的には、Web サービスの提供を想定し、その実現に必要な知識・技術を解説する。
併せて、具体的なサービス構築を通じ知識・技術の活用に加え、それらの獲得方法を実践的に体得する。(シラバスより)
